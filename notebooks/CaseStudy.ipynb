Ein zuverlässiger und pünktlicher öffentlicher Verkehr stellt einen zentralen Bestandteil der Mobilität in der Schweiz dar. Insbesondere der Bahnverkehr spielt dabei eine bedeutende Rolle, da täglich zahlreiche Menschen für berufliche, schulische oder private Zwecke auf die Züge der Schweizerischen Bundesbahnen (SBB) angewiesen sind. Die SBB gelten als eines der pünktlichsten Bahnunternehmen und leisten damit einen wichtigen Beitrag zur Funktionsfähigkeit des Schweizer Verkehrssystems.

Trotz der hohen Zuverlässigkeit des Bahnnetzes kann die Fahrplanstabilität durch verschiedene Faktoren beeinflusst werden. Neben betrieblichen Ursachen wie technischen Störungen, Bauarbeiten oder einer hohen Netzauslastung können auch äussere Einflüsse eine Rolle spielen. Dazu zählen unter anderem meteorologische Bedingungen, die potenziell Auswirkungen auf den Bahnbetrieb haben können.

Vor diesem Hintergrund beschäftigt sich dieses Projekt mit dem Thema Einfluss von Niederschlag auf die Fahrplanstabilität der Züge der SBB.



# Ziel

Ziel dieser Arbeit ist es, einen möglichen Zusammenhang zwischen Niederschlag und der Fahrplanstabilität der Züge der SBB zu untersuchen. Dabei soll analysiert werden, ob und in welchem Ausmass Niederschlagsereignisse mit Veränderungen in der Pünktlichkeit und Ausfällen des Bahnverkehrs in Verbindung stehen.

Zu diesem Zweck werden Wetterdaten zu Niederschlagsmengen mit verfügbaren Abfahrts- und Ankunftszeiten der SBB kombiniert und statistisch ausgewertet. Durch die Analyse dieser Daten soll geprüft werden, ob sich Hinweise auf einen Zusammenhang zwischen der Intensität von Niederschlag und möglichen Fahrplanabweichungen erkennen lassen.

Die Ergebnisse der Untersuchung sollen dazu beitragen, ein besseres Verständnis für mögliche wetterbedingte Einflüsse auf die Stabilität des Bahnverkehrs zu gewinnen.


## Vorgehen


Um nach der Einleitung und Zielsetzung die Fragestellungen im nächsten Kapitel zu beantworten, wird ein datengetriebener Analyseansatz verfolgt. Dieser orientiert sich an den Grundprinzipien der datengestützten Entscheidungsfindung (Data-Driven Decision Making). Den Prozess gliedern wir in mehrere aufeinanderfolgende Schritte.

1.   Datenbeschaffung: Zunächst werden die Daten beschafft. Die IST-Daten des Bahnverkehrs stammen aus dem Open-Data-Portal des öffentlichen Verkehrs, während die Niederschlagsdaten von MeteoSchweiz bezogen werden. Wir haben uns für diese Analyse bewusst für die Jahre 2023 und 202, sowie der Regionen Zürich und Bern entschieden, da sie eine gute Grundlage für unsere Untersuchungen bieten. Die beiden Datensätze dienen als Basis für alle weiteren Analysen.

2.   Datenaufbereitung: In einem zweiten Schritt erfolgt dann die Aufbereitung und Bereinigung der Daten. Hierzu zählt das Eliminieren unvollständiger oder falscher Einträge, das Vereinheitlichen von Zeitformaten und das Herausfiltern relevanter Variablen, also die Datentransformation. Die Fahrplandaten werden vor allem genutzt, um die effektiven Verspätungen in Minuten zu ermitteln, indem die geplanten und tatsächlichen Ankunfts- beziehungsweise Abfahrtszeiten miteinander gegenübergestellt werden.

3.   Datenverknüpfung: Im dritten Schritt werden die Datensätze miteinander verknüpft. Hierbei werden die Daten des Fahrplans mit den Wetterdaten kombiniert. Fahrten werden vorrangig anhand ihres zeitlichen und örtlichen Bezuges zugeordnet. Hierfür werden den entsprechenden Bahnhöfen oder Streckenabschnitten die nächstgelegenen Wetterstationen zugewiesen. Im Anschluss daran werden die Niederschlagswerte in geeigneten Zeitintervallen, wie etwa stündlich oder täglich, den entsprechenden Fahrten zugeordnet. Hierdurch wird ein kombinierter Datensatz generiert, der betriebliche und meteorologische Informationen umfasst.

4.   Datenanalyse: Im vierten Schritt wird dann eine explorative Datenanalyse vorgenommen. Die Absicht besteht darin, allfällige Muster, Auffälligkeiten und potenzielle Verbindungen zwischen Niederschlag und Verspätungen zu erkennen. Dies geschieht unter anderem durch statistische Kennzahlen, Visualisierungen und den Vergleich von Tagen mit und ohne Niederschlag.

5.   Datenauswertung: Im fünften Schritt widmen wir uns der statistisch‑analytischen Auswertung. Hierzu werden geeignete Methoden verwendet, um den Zusammenhang zwischen Niederschlagsmenge und Verspätungen quantitativ zu analysieren. Darunter fallen unter anderem Korrelationsanalysen und die Analyse von Schwellenwerten, ab denen signifikante Veränderungen in der Pünktlichkeit zu beobachten sind. Um ein differenzierteres Bild zu erhalten, können bei Bedarf weitere Einflussfaktoren wie Temperatur oder Zugtyp in die Analyse aufgenommen werden.

6.   Interpretation und Bewertung: Im letzten Schritt erfolgt die Interpretation und Bewertung der Ergebnisse. Die gewonnenen Erkenntnisse werden im Hinblick auf die Fragestellung erörtert und kritisch eingeordnet. Im Fokus steht vor allem die Frage, ob ein statistisch bedeutsamer Zusammenhang zwischen Niederschlag und Fahrplanabweichungen besteht und welche praktischen Konsequenzen sich daraus für den Bahnbetrieb ergeben.

Mit diesem klaren Vorgehen können wir nachvollziehbar und fundiert analysieren, ob und wie sich Niederschlag auf die Pünktlichkeit der SBB auswirkt und somit die nachfolgenden Fragestellungen klar beantworten.

# Datengrundlagen

Die vorliegende Analyse stützt sich auf zwei zentrale Datenquellen, welche die Grundlage für die Untersuchung des Zusammenhangs zwischen Niederschlagsereignissen und der Fahrplanstabilität der Schweizerischen Bundesbahnen (SBB) bilden. Um die Korrelation zu untersuchen, werden die täglichen Daten der Jahre 2023 und 2024 verglichen.

Diese Analyse stützt sich auf diese beiden Jahre, da diese im Archiv von Open Transport Data Swiss bereitgestellt sind und die täglichen Daten vorliegen.

**Ist-Daten des Schweizer Bahnverkehrs**

Die Fahrplandaten stammen aus dem Open-Data-Portal des öffentlichen Verkehrs (data.opentransportdata.swiss). Der dort bereitgestellte Datensatz «Ist-Daten Archiv» enthält tagesaktuelle Informationen über alle Fahrten im schweizerischen öffentlichen Verkehr. Für jede Fahrt werden unter anderem folgende Attribute erfasst: die geplante und tatsächliche Ankunfts- sowie Abfahrtszeit an den jeweiligen Haltestellen, die Linien- und Fahrtbezeichnung sowie allfällige Ausfälle oder Streckenänderungen. Aus diesen Rohdaten lässt sich die Verspätung einzelner Züge in Minuten ableiten, wodurch eine quantitative Bewertung der Pünktlichkeit auf Fahrt-, Strecken- und Tagesebene möglich wird.

**Niederschlagsdaten von MeteoSchweiz**

Die meteorologischen Daten werden vom nationalen Wetter- und Klimadienst MeteoSchweiz bezogen. Dabei kommen zwei komplementäre Messnetze zum Einsatz:
Das automatische Messnetz (ANETZ) liefert kontinuierliche, hochaufgelöste Niederschlagsmessungen im Stundenraster. Die Stationen messen den Niederschlag vollautomatisch und in Echtzeit, was eine zeitlich präzise Zuordnung von Wetterereignissen zu den entsprechenden Fahrten ermöglicht.
Ergänzend dazu werden Daten des manuellen Messnetzes herangezogen, welches tägliche Niederschlagssummen auf Basis von Ablesungen durch geschulte Beobachterinnen und Beobachter bereitstellt. Dieses Netz zeichnet sich durch eine langjährige, homogene Datenreihe aus und erlaubt eine räumlich breitere Abdeckung, insbesondere in abgelegenen Regionen der Schweiz.
Durch die Kombination beider Messnetze wird eine möglichst vollständige geografische und zeitliche Abdeckung der Niederschlagssituation in der Schweiz sichergestellt.
